```python
from requests import get
import pandas as pd
import json
```

```python
api_url = 'https://www.dnd5eapi.co'
```

```python
resp = get(api_url + '/api/monsters')
data = json.loads(resp.content)
data
```

```python
monsters = pd.DataFrame({
    'name': [mon['name'] for mon in data['results']],
    'url': [mon['url'] for mon in data['results']]
})
monsters
```

```python
mons = [
    json.loads(get(api_url+mon_url).content)
    for mon_url in monsters.loc[:, 'url']
]
```

```python
mns = pd.DataFrame(mons)
mns
```
